```python
#Discover outliers with visualzation tools
#Create a boxplot for 'arr_delay'
import seaborn as sns
sns.boxplot(x=df['arr_delay'])
```

```python
#Get Z-scores of flights data
from scipy import stats
import numpy as np
z = np.abs(stats.zscore(df))
print(z)
```

```python
# Define a threshold to identify an outlier.
threshold = 3
print(np.where(z > 3))
```

```python
print(z[100][2])
```

```python
#Find outliers using IQR Score(i.e. IQR=Q3-Q1) 
Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1
print(IQR)
```

```python
print(df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))
```

```python
#Get clean data by removing or filtering the outliers
flight_df_o = df[(z < 3).all(axis=1)]
```

```python
#Use IQR score to filterout the outliers to keep only valid values
df = df[~((df < (Q1 - 1.5 * IQR)) |(df > (Q3 + 1.5 * IQR))).any(axis=1)]
df.shape
```
